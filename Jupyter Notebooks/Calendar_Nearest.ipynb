```python
def dist(lat1, lng1, lat2, lng2):
    
    distance = hs.haversine((lat1, lng1), (lat2, lng2))
    
    return round(distance, 2)
```

```python
def nearest(circuits, parsed):
    start = parsed[-1]
    min_dist = float("inf")
    
    for circuit in circuits:
        if circuit in parsed:
            continue
        
        lat1 = f1_23.loc[f1_23['circuitRef'] == start, 'lat'].values[0]
        lng1 = f1_23.loc[f1_23['circuitRef'] == start, 'lng'].values[0]
        
        lat2 = f1_23.loc[f1_23['circuitRef'] == circuit, 'lat'].values[0]
        lng2 = f1_23.loc[f1_23['circuitRef'] == circuit, 'lng'].values[0]
        
        distance = dist(lat1, lng1, lat2, lng2)
        
        if distance < min_dist:
            min_dist = distance
            next_loc = circuit
        
    return(next_loc, min_dist)       
```

```python
def furthest(locations, parsed):
    start = parsed[-1]
    max_dist = 0
    
    for location in locations:
        if location in parsed:
            continue
        
        lat1 = f1_23.loc[f1_23['location'] == start, 'lat'].values[0]
        lng1 = f1_23.loc[f1_23['location'] == start, 'lng'].values[0]
        
        lat2 = f1_23.loc[f1_23['location'] == location, 'lat'].values[0]
        lng2 = f1_23.loc[f1_23['location'] == location, 'lng'].values[0]
        
        distance = dist(lat1, lng1, lat2, lng2)
        
        if distance > max_dist:
            max_dist = distance
            next_loc = location
        
    return(next_loc, max_dist)       
```

```python
def furthest_calendar(start, display=False):
    parsed = list()

    #start = locations[0]
    parsed.append(start)

    total = 0
    
    if display == True:
        print(start, 0)
    
    for location in locations:
    
        #print(parsed, "\n")
    
        if len(parsed) != len(locations):
            next_loc, distance = furthest(locations, parsed)
            parsed.append(next_loc)
            total = total + distance
            
            if display == True:
                print(next_loc, distance, "kms")
    
    if display == False:
        return parsed, total
    else:
        print("\nTotal distance :", round(total, 2), "kms")
```

```python
def loc_rank(start, locations):
    
    lat1 = f1_23.loc[f1_23['location'] == start, 'lat'].values[0]
    lng1 = f1_23.loc[f1_23['location'] == start, 'lng'].values[0]
    
    rank = {}
    
    
    for location in locations:
        lat2 = f1_23.loc[f1_23['location'] == location, 'lat'].values[0]
        lng2 = f1_23.loc[f1_23['location'] == location, 'lng'].values[0]
        
        distance = dist(lat1, lng1, lat2, lng2)
        
        rank[location] = distance
    
    ranked = dict(sorted(rank.items(), key=lambda item: item[1]))
    #s = ranked.pop(start)
    
    #print(ranked)
    return ranked
```

```python
def dist_rank(start, loc, parsed, display=False):
    loc_list = list(set(locations) - set(parsed))
    
    rank = loc_rank(start, loc_list)
    
    if display == True:
        print("\n", rank, "\n")
    rank_list = list(rank.keys())
    
    for i in range(0, len(rank_list)):
        if rank_list[i] == loc:
            return i+1, len(rank_list)    
```
