```python
def dist_calc(year, disp=False):
    cal = race_info.loc[race_info['year'] == year]
    
    rounds = cal.shape[0]
    lat1 = cal.loc[cal['round'] == 1, 'lat'].values[0]
    lng1 = cal.loc[cal['round'] == 1, 'lng'].values[0]
    loc1 = cal.loc[cal['round'] == 1, 'location'].values[0]

    
    total = 0

    for index, row in cal.iterrows():
        #print(row['circuitRef'])
        r = row['round']
        if r>1:
            lat2 = row['lat']
            lng2 = row['lng']
            loc2 = row['location']
        
            dist = hs.haversine((lat1, lng1), (lat2, lng2))
            
            if disp==True:
                print(loc1, '<->', loc2, "=", round(dist, 2), "kms\n")
            total = total + dist
        
            lat1 = lat2
            lng1 = lng2
            loc1 = loc2
        
            #print(r)
    if disp==True:
        print("-----------------------------------------")
    
        print("\nF1 {0} Total Distance : {1} kilometres".format(year, round(total, 2)))
    
    return round(total, 2), rounds
```

```python
def header_count(year):
    season = race_info.loc[race_info['year'] == year]
    
    single = double = triple = 0
    double_dup = triple_dup = False
    
    for index, row in season.iterrows():
        if row['gap'] > 1:
            double_dup = triple_dup = False
            
        if row['header'] == 'single':
            single = single + 1
        elif row['header'] == 'double':
            if double_dup == False:
                double = double + 1
            double_dup = True
        elif row['header'] == 'triple':
            if triple_dup == False:
                triple = triple + 1
            triple_dup = True

    
    return single, double, triple 
```

```python
def header_dist(year, disp=False):
    df = race_info.loc[race_info['year'] == year]
    
    double_flag = triple_flag1 = triple_flag2 = False
    double_dist = triple_dist = 0
    
    for index, row in df.iterrows():
        if double_flag == True:
            double_dist = double_dist + row['distance (km)']
            if disp == True:
                print('Double', row['circuitRef'], row['distance (km)'])
            double_flag = False
            continue
            
        if triple_flag1 == True:
            triple_dist = triple_dist + row['distance (km)']
            if disp == True:
                print('Triple 1', row['circuitRef'], row['distance (km)'])
            triple_flag2 = True
            triple_flag1 = False
            continue
            
        if triple_flag2 == True:
            triple_dist = triple_dist + row['distance (km)']
            if disp == True:
                print('Triple 2', row['circuitRef'], row['distance (km)'])
            triple_flag1 = triple_flag2 = False
            continue
                 
        if row['header'] == 'double':
            double_flag = True  
            
        if row['header'] == 'triple':
            triple_flag1 = True
            
    double_dist = round(double_dist, 2)
    triple_dist = round(triple_dist, 2)
    if disp == True:
        print("\nDouble distance :", double_dist)
        print("Triple distance :", triple_dist)
        
    return double_dist, triple_dist
```
