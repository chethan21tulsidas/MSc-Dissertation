```python
carbon = pd.read_csv("Carbon_Footprint.csv")

carbon.shape
```

```python
carbon.head(10)
```

```python
carbon['year'].unique().tolist()
```

```python
cols = ['year', 'total distance (km)', 'footprint (TCO2e)']
row = dict.fromkeys(cols, 0)
footprint = pd.DataFrame(columns = cols)

for year in carbon['year'].unique().tolist():
    dist = carbon.loc[carbon['year'] == year, 'distance (km)'].sum()
    fp = carbon.loc[carbon['year'] == year, 'footprint'].sum()
    
    print(year, round(dist, 2), round(fp, 2))
    
    row['year'] = year
    row['total distance (km)'] = round(dist, 2)
    row['footprint (TCO2e)'] = round(fp, 2)
    
    cfp = pd.DataFrame([row])
    
    footprint = pd.concat([footprint, cfp], ignore_index=True)
```

```python
footprint
```

```python
footprint.to_csv("Season_Footprint.csv", index=False)
```
