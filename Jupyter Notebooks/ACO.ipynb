```python
def calculate_probabilities(current_location, unvisited_locations):
    probabilities = []
    total_prob = 0
    for loc in unvisited_locations:
        pheromone_factor = pheromones[current_location, loc] ** alpha
        heuristic_factor = heuristic_info[current_location, loc] ** beta
        probability = pheromone_factor * heuristic_factor
        probabilities.append(probability)
        total_prob += probability
    probabilities = [p / total_prob for p in probabilities]
    return probabilities
```

```python
def adjust_probabilities(current_location, unvisited_locations, race):
    probabilities = []
    total_prob = 0
    for loc in unvisited_locations:
        pheromone_factor = pheromones[current_location, loc] ** alpha
        heuristic_factor = heuristic_info[current_location, loc] ** beta
        if race not in windows[circuits[loc]]:
            probability = 0
        else:
            probability = pheromone_factor * heuristic_factor
        probabilities.append(probability)
        total_prob += probability
    if total_prob == 0:
        probabilities = [0 for p in probabilities]
    else:
        probabilities = [p / total_prob for p in probabilities]
    return probabilities    
```

```python
num_ants = 40
num_iterations = 500
evaporation_rate = 0.1
initial_pheromone = 0.1
pheromone_intensity = 1.0
pheromone_deposit_limit = 1.0
alpha = 1  # Pheromone factor
beta = 2   # Heuristic factor

pheromones = np.full_like(distance_matrix, initial_pheromone)
heuristic_info = 1 / distance_matrix

best_solution = None
best_distance = float('inf')

for _ in range(num_iterations):
    solutions = []

    for ant in range(num_ants):
        unvisited_locations = list(range(len(distance_matrix)))
        current_location = np.random.choice(unvisited_locations)
        unvisited_locations.remove(current_location)
        solution = [current_location]
        race = 1
        
        while unvisited_locations:
            #probabilities = calculate_probabilities(current_location, unvisited_locations)
            
            race += 1
            probabilities = adjust_probabilities(current_location, unvisited_locations, race)
            if all(p == 0 for p in probabilities):
                break
            
            next_location = np.random.choice(unvisited_locations, p=probabilities)
            solution.append(next_location)
            unvisited_locations.remove(next_location)
            current_location = next_location
        
        if len(solution) == 24:
            solutions.append(solution)
    
    
    # Evaluate solutions and update pheromones
    for solution in solutions:
        distance = sum(distance_matrix[solution[i], solution[i+1]] for i in range(len(solution) - 1))
        weather_windows_satisfied = all(race in windows[circuits[location]] for location, race in zip(solution, range(1, 25)))

        """
        if weather_windows_satisfied and distance < best_distance:
            best_distance = distance
            best_solution = solution
            print(_, best_solution, distance)
        """
        
        if distance < best_distance:
            best_distance = distance
            best_solution = solution
            print(_, best_solution, distance)
        
        for i in range(len(solution) - 1):
            pheromones[solution[i], solution[i+1]] = min(pheromones[solution[i], solution[i+1]] + pheromone_intensity / distance, pheromone_deposit_limit)
            pheromones[solution[i+1], solution[i]] = pheromones[solution[i], solution[i+1]]
    
    # Evaporate pheromones
    pheromones *= (1 - evaporation_rate)
    
    
```
