```python
f1_weather_30years = pd.merge(f1_30years, weather_30years, on=['circuitRef', 'year', 'month'], how='left')

f1_weather_30years
```

```python
cols = ['temp_mean (°C)', 'temp_max (°C)', 'temp_min (°C)', 'precipitation_mean (mm)', 'rain_mean (mm)', 'snow_mean (cm)', 'precipitation_total (mm)', 'rain_total (mm)', 'snow_total (cm)', 'moderate_days (>25.55°C)', 'high_days (>27.77°C)', 'precipitation_days', 'rain_days', 'snow_days']

f1_weather_30years[cols].describe()
```

```python
for col in cols:
    high = f1_weather_30years[col].max()
    
    high_race = f1_weather_30years.loc[f1_weather_30years[col] == high, ['circuitRef', 'year', col]].values
    
    print(col, high)
    print(high_race)
    print()
```

```python
for col in cols:
    
    q1 = round(f1_weather_30years[col].quantile(.25), 2)
    q3 = round(f1_weather_30years[col].quantile(.75), 2)
    iqr = round(q3-q1, 2)
    
    high = round(f1_weather_30years[col].max(), 2)
    mean = round(f1_weather_30years[col].mean(), 2)
    
    std_3 = round(f1_weather_30years[col].std() * 3, 2)
    high_outlier = round(mean + std_3, 2)
    low_outlier = round(mean - std_3, 2)
    

    
    print(col)
    print("IQR", q3, q1, iqr, round((q3 + (1.5*iqr)), 2), round((q1 - (1.5*iqr)), 2), high)
    print("Std Dev.", mean, std_3, high_outlier, low_outlier)
    
    for index, row in f1_weather_30years.iterrows():
        if row[col] > (q3 + (1.5*iqr)):
            print("IQR High outlier :", row['circuitRef'], row['year'], row[col])
                       
        if row[col] < (q1 - (1.5*iqr)):
            print("IQR Low outlier :", row['circuitRef'], row['year'], row[col])
        
        if row[col] > high_outlier:
            print("Std Dev. High Outlier :", row['circuitRef'], row['year'], row[col])
        
        if row[col] < low_outlier:
            print("Std Dev. Low Outlier :", row['circuitRef'], row['year'], row[col])
    
    print()
```

```python
optimal_cols = ['circuitRef', 'month', 'temp_mean (°C)', 'temp_max (°C)', 'temp_min (°C)', 'precipitation_mean (mm)', 'rain_mean (mm)', 'snow_mean (cm)', 'precipitation_total (mm)', 'rain_total (mm)', 'snow_total (cm)', 'moderate_days (>25.55°C)', 'high_days (>27.77°C)', 'precipitation_days', 'rain_days', 'snow_days']

optimal_weather = pd.DataFrame(columns = optimal_cols)

for index, row in average_weather.loc[average_weather['circuitRef'].isin(f1_23_tracks)].iterrows():
    
    optimal_info = dict.fromkeys(optimal_cols, 0)
    
    optimal_info['circuitRef'] = row['circuitRef']
    optimal_info['month'] = row['month']
    
    for col in cols:
    
        q1 = round(f1_weather_30years[col].quantile(.25), 2)
        q3 = round(f1_weather_30years[col].quantile(.75), 2)
        iqr = round(q3-q1, 2)
        
        mean = round(f1_weather_30years[col].mean(), 2)
        std_3 = round(f1_weather_30years[col].std() * 3, 2)
        high_outlier = round(mean + std_3, 2)
        low_outlier = round(mean - std_3, 2)
        
        if (row['month'] < 3) | (row['month'] > 11):
            optimal_info[col] = 'Off-season'
            continue
        
        """
        if row[col] <= (q3 + (1.5*iqr)):
            if row[col] >= (q1 - (1.5*iqr)):
                #print("Optimal :", row['circuitRef'], row['month'],row[col])
                optimal_info[col] = "Optimal"
                
        if row[col] > (q3 + (1.5*iqr)):
            #print("High outlier :", row['circuitRef'], row['month'],row[col])
            optimal_info[col] = "High outlier"
                       
        if row[col] < (q1 - (1.5*iqr)):
            #print("Low outlier :", row['circuitRef'], row['month'], row[col])
            optimal_info[col] = "Low outlier"
        """
        
        if row[col] < high_outlier:
            if row[col] > low_outlier:
                #print("Optimal :", row['circuitRef'], row['month'],row[col])
                optimal_info[col] = "Optimal"
        
        if row[col] >= high_outlier:
            #print("High outlier :", row['circuitRef'], row['month'],row[col])
            optimal_info[col] = "High outlier"
        
        if row[col] <= low_outlier:
            #print("Low outlier :", row['circuitRef'], row['month'], row[col])
            optimal_info[col] = "Low outlier"
            
    
    optimal_row = pd.DataFrame([optimal_info])
        
    optimal_weather = pd.concat([optimal_weather, optimal_row], ignore_index=True)
    
```
